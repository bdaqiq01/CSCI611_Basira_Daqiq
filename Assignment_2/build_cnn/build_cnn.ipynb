## Load the [Data](http://pytorch.org/docs/stable/torchvision/datasets.html)
---
Downloading may take a minute. We load in the training and test data, split the training data into a training and validation set, then create DataLoaders for each of these sets of data.

```python
from torchvision import datasets
import torchvision.transforms as transforms
from torch.utils.data.sampler import SubsetRandomSampler

# number of subprocesses to use for data loading
num_workers = 0
# how many samples per batch to load
batch_size = 20
# percentage of training set to use as validation
valid_size = 0.2

# convert data to a normalized torch.FloatTensor
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])

# choose the training and test datasets
train_data = datasets.CIFAR10('data', train=True,
                              download=True, transform=transform)
test_data = datasets.CIFAR10('data', train=False,
                             download=True, transform=transform)

# obtain training indices that will be used for validation
num_train = len(train_data)
indices = list(range(num_train))
np.random.shuffle(indices)
split = int(np.floor(valid_size * num_train))
train_idx, valid_idx = indices[split:], indices[:split]

# define samplers for obtaining training and validation batches
train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)

# prepare data loaders (combine dataset and sampler)
train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size,
    sampler=train_sampler, num_workers=num_workers)
valid_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size, 
    sampler=valid_sampler, num_workers=num_workers)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=batch_size, 
    num_workers=num_workers)

# specify the image classes
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck']
```

## **TODO**: Define the Network [Architecture](http://pytorch.org/docs/stable/nn.html)

Build up your own Convolutional Neural Network using Pytorch API:
- nn.Conv2d(): for convolution
- nn.MaxPool2d(): for maxpooling (spatial resolution reduction)
- nn.Linear(): for last 1 or 2 layers of fully connected layer before the output layer.
- nn.Dropout(): optional, [dropout](https://pytorch.org/docs/stable/generated/torch.nn.Dropout.html) can be used to avoid overfitting.
- F.relu(): Use ReLU as the activation function for all the hidden layers

The following is a skeleton example that's not completely working. 

```python
import torch.nn as nn
import torch.nn.functional as F

# define the CNN architecture
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        
        # TOTO: Build multiple convolutional layers (sees 32x32x3 image tensor in the first hidden layer)
        # for example, conv1, conv2 and conv3
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1) #input 3*32*32 output: 16*32*32 #1 padding and stride 1
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1) #input 16*32*32    output: 32*32*32 #1 padding and stride 1
        self.conv3 = nn.Conv2d(32, 64, 3, padding=1) #input 32*32*32 output: 64*32*32 #1 padding and stride 1
        
        # max pooling layer
        self.pool = nn.MaxPool2d(2, 2)
        
        # DONE: Build some linear layers (fully connected)
        # for example, fc1 and fc2
        self.fc1 = nn.Linear(64*4*4, 100) #
        self.fc2 = nn.Linear(100, 10)

        # TODO: dropout layer (p=0.25, you can adjust)
        self.dropout = nn.Dropout(0.25)

    def forward(self, x):
        # add sequence of convolutional and max pooling layers
        # assume we have 2 convolutional layers defined agove
        # and we do a maxpooling after each conv layer
        x1 = F.relu(self.conv1(x)) #input 3*32*32 output: 16*32*32 after maxpooling 16*16*16
        x = self.pool(x1)
        x = self.pool(F.relu(self.conv2(x))) #input 16*16*16 output: 32*16*16 after maxpooling 32*8*8
        x = self.pool(F.relu(self.conv3(x)))  # input 32*8*8    output: 64*8*8 after maxpooling 64*4*4
        x = x.view(-1, 64*4*4)  #double check this number

        # optional add dropout layer
        x = self.dropout(x) #drops 25% of the neurons

        x = F.relu(self.fc1(x)) #input 64*4*4 output: 100*1*1 after relu
        # add 2nd hidden layer, with relu activation function
        x = self.dropout(x) #drops 25% of the neurons
        x = self.fc2(x) #input 100*1*1 output: 10*1*1
        return x1, x

# create a complete CNN
model = Net()
print(model)

# move tensors to GPU if CUDA is available
if train_on_gpu:
    model.cuda()
```

## Train the Network
---
Remember to look at how the training and validation loss decreases over time; if the validation loss ever increases it indicates possible overfitting.

The following is working code, but you are encouraged to make your own adjustments and enhance the implementation.

```python
# number of epochs to train the model, you decide the number
n_epochs = 15

valid_loss_min = np.inf # track change in validation loss

loss_dt =pd.DataFrame(columns=['epoch', 'train_loss', 'valid_loss'])

for epoch in range(1, n_epochs+1):

    # keep track of training and validation loss
    train_loss = 0.0
    valid_loss = 0.0
    
    ###################
    # train the model #
    ###################
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        # move tensors to GPU if CUDA is available
        if train_on_gpu:
            data, target = data.cuda(), target.cuda()
        # clear the gradients of all optimized variables
        optimizer.zero_grad()
        # forward pass: compute predicted outputs by passing inputs to the model
        output1, output = model(data)
        # calculate the batch loss
        loss = criterion(output, target)
        # backward pass: compute gradient of the loss with respect to model parameters
        loss.backward()
        # perform a single optimization step (parameter update)
        optimizer.step()
        # update training loss
        train_loss += loss.item()*data.size(0)
        
    
    ######################    
    # validate the model #
    ######################
    valid_loss = 0.0
    correct = 0
    total = 0

    model.eval()
    for batch_idx, (data, target) in enumerate(valid_loader):
        # move tensors to GPU if CUDA is available
        if train_on_gpu:
            data, target = data.cuda(), target.cuda()
        # forward pass: compute predicted outputs by passing inputs to the model
        output1, output = model(data)
        # calculate the batch loss
        loss = criterion(output, target)
        # update average validation loss 
        valid_loss += loss.item()*data.size(0)
    
    # calculate average losses
    train_loss = train_loss/len(train_loader.sampler)
    valid_loss = valid_loss/len(valid_loader.sampler)
    
    new_row = pd.DataFrame({'epoch': [epoch], 'train_loss': [train_loss], 'valid_loss': [valid_loss]})
    loss_dt = pd.concat([loss_dt, new_row], ignore_index=True)
    #train_acc = 
        
    # print training/validation statistics 
    print('Epoch: {} \tTraining Loss: {:.6f} \tValidation Loss: {:.6f}'.format(
        epoch, train_loss, valid_loss))
    
    # save model if validation loss has decreased
    if valid_loss <= valid_loss_min:
        print('Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
        valid_loss_min,
        valid_loss))
        torch.save(model.state_dict(), 'model_trained1.pt')
        valid_loss_min = valid_loss
```

## Test the Trained Network
---
Test your trained model on previously unseen data! Remember we have downloaded `train_data` and `test_data`. We will use `test_data` through `test_loader`.

A "good" result will be a CNN that gets around 70% (or more, try your best!) accuracy on these test images.

```python
# track test loss
test_loss = 0.0
class_correct = list(0. for i in range(10))
class_total = list(0. for i in range(10))

selected_channel = [0, 5, 10]
top_k = 5
top_images = {f: [] for f in selected_channel}
global_img_index = 0 


model.eval()
# iterate over test data
for batch_idx, (data, target) in enumerate(test_loader):
    # move tensors to GPU if CUDA is available
    if train_on_gpu:
        data, target = data.cuda(), target.cuda()
    # forward pass: compute predicted outputs by passing inputs to the model
    output1, output = model(data)
    batch_size = data.size(0)   
    activation = output1.view(batch_size, 16,-1).max(dim=2)[0]
    #activation = activation.cpu().detach().numpy()
    for i in range(batch_size):
        for f in selected_channel:
            ac_val = activation[i, f].item()
            top_images[f].append((ac_val, data[i].cpu().numpy(), target[i].item(), global_img_index))
        global_img_index += 1
    # calculate the batch loss
    loss = criterion(output, target)
    # update test loss 
    test_loss += loss.item()*data.size(0)
    # convert output probabilities to predicted class
    _, pred = torch.max(output, 1)    
    # compare predictions to true label
    correct_tensor = pred.eq(target.data.view_as(pred))
    correct = np.squeeze(correct_tensor.numpy()) if not train_on_gpu else np.squeeze(correct_tensor.cpu().numpy())
    # calculate test accuracy for each object class
    for i in range(data.size(0)):  # use actual batch size instead of fixed batch_size
        label = target[i].item()  # use .item() to get Python int
        class_correct[label] += correct[i].item()
        class_total[label] += 1

# ----- Keep only top_k images per filter -----
for f in selected_channel:
    # Sort by activation (highest first)
    top_images[f] = sorted(top_images[f], key=lambda x: x[0], reverse=True)[:top_k]

# average test loss
test_loss = test_loss/len(test_loader.dataset)
print('Test Loss: {:.6f}\n'.format(test_loss))

for i in range(10):
    if class_total[i] > 0:
        print('Test Accuracy of %5s: %2d%% (%2d/%2d)' % (
            classes[i], 100 * class_correct[i] / class_total[i],
            int(class_correct[i]), int(class_total[i])))
    else:
        print('Test Accuracy of %5s: N/A (no training examples)' % (classes[i]))

print('\nTest Accuracy (Overall): %2d%% (%2d/%2d)' % (
    100. * sum(class_correct) / sum(class_total),
    int(sum(class_correct)), int(sum(class_total))))
```

# Feature map from the first conv layer

```python
def plot_feature_maps(feature_maps, selected_images_np, selected_label=None):
    """
    Plot feature maps in a 4x4 grid for a single image.
    
    Parameters:
    - feature_maps: numpy array of shape (n_filters, height, width) - single image's feature maps
    - selected_images_np: numpy array of shape (3, height, width) - single original image
    - selected_label: optional class label (int) for title
    """

    
    fig = plt.figure(figsize=(20, 20))
    
    # Calculate spacing: 1 inch border + 1 inch between grid items
    # Figure is 20x20, so 1 inch = 1/20 = 0.05 in figure coordinates
    border = 1.0  # 1 inch border
    spacing = 1.0  # 1 inch spacing between items
    
    # Calculate grid dimensions
    grid_width = 20 - 2 * border
    grid_height = 20 - 2 * border
    
    # Size of each item in the 4x4 grid
    # 4 items + 3 spacings = grid_width
    item_width = (grid_width - 3 * spacing) / 4
    item_height = (grid_height - 3 * spacing) / 4
    
    # Starting position for the grid (top-left)
    grid_top = 20 - border
    grid_left = border
    
    # Plot original image in position (0,0) of the 4x4 grid
    ax_left = grid_left
    ax_bottom = grid_top - item_height
    ax = fig.add_axes([ax_left/20, ax_bottom/20, item_width/20, item_height/20])
    ax.imshow(np.transpose(selected_images_np, (1, 2, 0)) / 2 + 0.5)  # unnormalize
    if selected_label is not None:
        ax.set_title(f'Original\n{selected_label}', fontsize=10, fontweight='bold')
    else:
        ax.set_title('Original', fontsize=10, fontweight='bold')
    ax.axis('off')
    
    # Plot 15 feature maps in remaining positions
    filter_idx = 0
    for row in range(4):
        for col in range(4):
            if row == 0 and col == 0:
                continue  # Skip position (0,0) - that's the original image
            
            if filter_idx >= 15:
                break
            
            # Calculate position for this filter
            ax_left = grid_left + col * (item_width + spacing)
            ax_bottom = grid_top - (row + 1) * (item_height + spacing)
            
            ax = fig.add_axes([ax_left/20, ax_bottom/20, item_width/20, item_height/20])
            ax.imshow(feature_maps[filter_idx], cmap='viridis')
            ax.set_title(f' Con2D layer 1, Filter {filter_idx}', fontsize=8)
            ax.axis('off')
            
            filter_idx += 1
    
    plt.show() 
```
